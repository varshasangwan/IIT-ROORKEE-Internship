```python
import csv

# Writing student data to CSV file
with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Header row
    writer.writerow(["Roll No", "Name", "Marks"])

    # Student records
    writer.writerow([101, "Rahul", 85])
    writer.writerow([102, "Priya", 92])
    writer.writerow([103, "Amit", 78])

print("Student data written successfully.")

# Reading data from CSV file
print("\nStudent Records:")

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result =", result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
print(safe_divide(10,2))
print(safe_divide(10,0))
```

```python
def add_task(task):
    with open("todo.txt", "a") as file:
        file.write(task + "\n")
    print(f"Added: '{task}'")

def view_tasks():
    print("\n--- To-Do List ---")
    try:
        with open("todo.txt", "r") as file:
            tasks = file.readlines()

            if not tasks:
                print("No tasks found!")
            else:
                for i, task in enumerate(tasks, 1):
                    print(f"{i}. {task.strip()}")

    except FileNotFoundError:
        print("No to-do list found!")

    print("------------------")

# Running the app
add_task("Complete assignment")
add_task("Study Python")

view_tasks()
```

```python
with open("log.txt", "w") as file:
    file.write("ERROR\n")
    file.write("WARNING\n")
    file.write("ERROR\n")
    file.write("INFO\n")
    file.write("ERROR\n")
    file.write("WARNING\n")

print("Log file created!")
error_count = 0
warning_count = 0
info_count = 0

with open("log.txt", "r") as file:
    for line in file:
        if line.strip() == "ERROR":
            error_count += 1
        elif line.strip() == "WARNING":
            warning_count += 1
        elif line.strip() == "INFO":
            info_count += 1

print("ERROR =", error_count)
print("WARNING =", warning_count)
print("INFO =", info_count)
```

```python
import csv

with open("marks.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Marks"])
    writer.writerow(["Varsha", 85])
    writer.writerow(["Rahul", 90])
    writer.writerow(["Aman", 75])

print("CSV file created!")
import csv

total = 0
count = 0

with open("marks.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip header

    for row in reader:
        total += int(row[1])
        count += 1

average = total / count

with open("result.txt", "w") as file:
    file.write("Average Marks = " + str(average))

print("Average =", average)
```
