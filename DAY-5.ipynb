```python
sentence = input("Enter a sentence: ")
words = sentence.split()

frequency = {}

for word in set(words):
    frequency[word] = words.count(word)

print("Word Frequency:")
print(frequency)
```

```python
text = input("Enter a word: ")

char_map = {}
length = len(text)

# store characters with reversed index
for i in range(length):
    char_map[i] = text[length - 1 - i]

# build reversed string from dictionary
reversed_text = ""

for i in range(length):
    reversed_text += char_map[i]

print("Reversed word:", reversed_text)

if text == reversed_text:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    grade = input("Enter grade: ")

    students[name] = {"Marks": marks, "Grade": grade}

print("\nStudent Report Card")
for name, details in students.items():
    print(f"{name}: Marks = {details['Marks']}, Grade = {details['Grade']}")
```

```python
text = input("Enter message: ")
shift = int(input("Enter shift value: "))

encoded = ""

for char in text:
    if char.isalpha():
        if char.islower():
            encoded += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
        else:
            encoded += chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
    else:
        encoded += char

print("Encoded message:", encoded)
```
