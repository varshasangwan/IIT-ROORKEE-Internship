```python
# 'UDI' aur 'Product ID' columns ko hatane ke liye (axis=1 matlab columns)
df_clean = df.drop(['UDI', 'Product ID'], axis=1)

print("New data columns:")
print(df_clean.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graph ka size set karne ke liye
plt.figure(figsize=(6,4))

# Machine Failure ka bar chart banane ke liye
sns.countplot(x='Machine failure', data=df_clean, hue='Machine failure', palette='Set2', legend=False)

# Graph m titles lagane ke liye
plt.title('Distribution of Machine Failure (0 = Normal, 1 = Failure)')
plt.xlabel('Machine Status')
plt.ylabel('Count')

# Graph ko screen par dikhane ke liye
plt.show()
```

```python
# 'Machine failure' column mein 0 aur 1 kitni baar hain, exact count dekhne ke liye
print(df_clean['Machine failure'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graph ka background style set karne ke liye
sns.set_theme(style="whitegrid")

# Graph ka size (Width=8, Height=5) set karne ke liye
plt.figure(figsize=(8, 5))

# Air temperature ka histogram aur bell curve (KDE) banane ke liye
# Note: Agar column ka 'A' small ho, toh 'air temperature [K]' likhiyega
sns.histplot(data=df_clean, x='Air temperature [K]', kde=True, color='teal', bins=30)

# Labels aur Title lagane ke liye
plt.title('Histogram of Air Temperature [K]', fontsize=14, fontweight='bold')
plt.xlabel('Air Temperature [K]', fontsize=12)
plt.ylabel('Frequency (Count)', fontsize=12)

# Graph ko screen par display karne ke liye
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn theme set karein (whitegrid bohot clean lagta hai)
sns.set_theme(style="whitegrid")

# Ek bada frame (Width=16, Height=5) aur subplots (1 row, 3 columns) banayein
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# --- Graph 1: Process temperature [K] ---
sns.histplot(ax=axes[0], data=df_clean, x='Process temperature [K]', kde=True, color='indianred', bins=30)
axes[0].set_title('Histogram of Process Temperature [K]', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Process Temperature [K]')
axes[0].set_ylabel('Frequency (Count)')

# --- Graph 2: Rotational speed [rpm] ---
sns.histplot(ax=axes[1], data=df_clean, x='Rotational speed [rpm]', kde=True, color='mediumpurple', bins=30)
axes[1].set_title('Histogram of Rotational Speed [rpm]', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Rotational Speed [rpm]')
axes[1].set_ylabel('Frequency (Count)')

# --- Graph 3: Tool wear [min] ---
# Tool wear m KDE nahi lagayenge kyunki yeh ek 'timer' hai, bell curve nahi.
sns.histplot(ax=axes[2], data=df_clean, x='Tool wear [min]', kde=False, color='goldenrod', bins=30)
axes[2].set_title('Histogram of Tool Wear [min]', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Tool Wear [min]')
axes[2].set_ylabel('Frequency (Count)')

# Subplots ke beech m space management (layout adjust karne ke liye)
plt.tight_layout()

# Saare graphs ko display karein
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))
# Tool wear ka distribution dekhna fail vs normal ke liye
sns.boxplot(x='Machine failure', y='Tool wear [min]', data=df_clean, palette='Set1', hue='Machine failure', legend=False)

plt.title('Tool Wear vs Machine Failure', fontsize=14, fontweight='bold')
plt.xlabel('Machine Failure (0 = Normal, 1 = Failed)', fontsize=12)
plt.ylabel('Tool Wear [min]', fontsize=12)
plt.show()
```
