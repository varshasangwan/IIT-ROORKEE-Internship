```python
def grade_tracker():
    # (Name, Marks)
    students = [ ("Amit", 85) ,("Riya", 92),("Rahul", 67),("Sneha", 78),("Vikas", 45)]
    print("Student Grade Tracker\n")    
    for name, marks in students:
        if marks >= 90:
            grade = "A+"
        elif marks >= 75:
            grade = "A"
        elif marks >= 60:
            grade = "B"
        elif marks >= 40:
            grade = "C"
        else:
            grade = "F"

        print("name-",name,";marks-",marks,";grades-",grade)

grade_tracker()
```

```python
A=[[1,2,3],[4,5,6]]
B=[[2,3,4],[4,6,5]]
result = [[0, 0, 0],[0, 0, 0]]

# Matrix addition using nested loops
for i in range(len(A)):          # rows
    for j in range(len(A[0])):   # columns
        result[i][j] = A[i][j] + B[i][j]

# Display result
print("Sum of matrices:")
for row in result:
    print(row)
```
