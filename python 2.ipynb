```python
def add(a, b):
 return a + b
add(3,4)
def sub(a,b):
 return a-b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero is not allowed.")
    return a / b
def average(a,b,c):
    return a+b+c/3
print(add(2,3))
print(sub(5,4))
print(multiply(3,6))
print(divide(6,3))
print(average(3,4,5))
```

```python
import random
def roll_dice(sides=6):
    return random.randint(1, sides)
def roll_multiple_dice(number_of_dice=2, sides=6):
     return [roll_dice(sides) for _ in range(number_of_dice)]
print(roll_dice())
print(roll_multiple_dice(4,6))
```

```python
def run_quiz():
    questions = [
        {
            "question": "What is the capital of India?",
            "options": ["A. Mumbai", "B. New Delhi", "C. Kolkata", "D. Chennai"],
            "answer": "B"
        },
        {
            "question": "What is 5 + 7?",
            "options": ["A. 10", "B. 11", "C. 12", "D. 13"],
            "answer": "C"
        },
        {
            "question": "Which planet is known as the Red Planet?",
            "options": ["A. Earth", "B. Mars", "C. Jupiter", "D. Venus"],
            "answer": "B"
        },
        {
            "question": "Which language is used for web development?",
            "options": ["A. Python", "B. Java", "C. HTML", "D. C++"],
            "answer": "C"
        },
        {
            "question": "What is the square of 4?",
            "options": ["A. 12", "B. 14", "C. 16", "D. 18"],
            "answer": "C"
        }
    ]

    score = 0
print("=== Welcome to the Quiz App ===\n")

for i, q in enumerate(questions, start=1):
        print(f"Q{i}: {q['question']}")
        
        for option in q["options"]:
            print(option)

        user_answer = input("Your answer (A/B/C/D): ").strip().upper()

        if user_answer == q["answer"]:
            print("Correct!\n")
            score += 1
        else:
            print(f"Wrong! Correct answer is {q['answer']}\n")

print("=== Quiz Finished ===")
print(f"Your Score: {score} / {len(questions)}")

if score == len(questions):
        print("Excellent! Perfect score!")
elif score >= 3:
    print("Good job!")
else:
    print("Keep practicing!")

if __name__ == "__main__":
    run_quiz()

```
