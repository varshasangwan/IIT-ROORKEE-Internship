```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Setup Sample Data
np.random.seed(42)
n = 500
df = pd.DataFrame({
    'Normal': np.random.normal(10, 2, n),
    'Right_Skewed': np.random.exponential(5, n),
    'Left_Skewed': 20 - np.random.exponential(3, n),
    'Group': np.random.choice(['A', 'B'], n)
})

# 2. Plot Histograms (Individual Distributions)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
colors = ['skyblue', 'salmon', 'lightgreen']
titles = ['Symmetrical (Normal)', 'Right-Skewed (Positive)', 'Left-Skewed (Negative)']

for i, col in enumerate(['Normal', 'Right_Skewed', 'Left_Skewed']):
    sns.histplot(df[col], kde=True, ax=axes[i], color=colors[i])
    axes[i].set_title(titles[i])

plt.tight_layout()
plt.show()

# 3. Plot Pair Plot (Relationships Between Features)
sns.pairplot(df, hue='Group', palette='Set2', diag_kind='kde')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species_name'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
print("--- 1. Shape of dataset ---")
print(df.shape)
print("\n--- 2. Statistical Summary (Describe) ---")
print(df.describe())
print("\n--- 3. Value Counts (how many flowers of respective species) ---")
print(df['species_name'].value_counts())
```
