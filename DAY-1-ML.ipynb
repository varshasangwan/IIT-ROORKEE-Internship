```python
import matplotlib.pyplot as plt

# Aapka data
ghar_size = [1000, 1500, 2000, 2500]
ghar_price = [50, 75, 100, 125]

# Graph banana
plt.plot(ghar_size, ghar_price, marker='o', color='green', linestyle='-')
plt.title("Ghar Ka Size vs Price")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price (Lakhs)")

# Graph ko Notebook ke andar hi dikhane ke liye
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species_name'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
print("--- 1. Shape of dataset ---")
print(df.shape)
print("\n--- 2. Statistical Summary (Describe) ---")
print(df.describe())
print("\n--- 3. Value Counts (how many flowers of respective species) ---")
print(df['species_name'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
df.drop(columns=['species', 'species_name']).hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions (Histograms)", fontsize=16)
plt.tight_layout()
plt.show()
sns.pairplot(df.drop(columns=['species']), hue='species_name', palette='Set2')
plt.suptitle("Iris Dataset Pair Plot", y=1.02, fontsize=16)
plt.show()
```
